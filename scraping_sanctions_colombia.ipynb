```python
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
# how to open the table that we need
driver.get("http://vital.minambiente.gov.co/SILPA_UT_PRE/RUIA/ConsultarSancion.aspx?Ubic=ext")
driver.find_element(By.XPATH, '/html/body/form/div[3]/div/main/div[3]/div[2]/div[2]/div[1]/div[2]/table[1]/tbody/tr[1]/td[2]/fieldset[2]/table/tbody/tr/td[1]/input').send_keys('01/01/2015')
driver.find_element(By.XPATH, '/html/body/form/div[3]/div/main/div[3]/div[2]/div[2]/div[1]/div[2]/table[1]/tbody/tr[1]/td[2]/fieldset[2]/table/tbody/tr/td[2]/input').send_keys('14/02/2022')
driver.find_element(By.XPATH, '/html/body/form/div[3]/div/main/div[3]/div[2]/div[2]/div[1]/div[2]/table[1]/tbody/tr[2]/td/table/tbody/tr[1]/td[2]/input').click()


```

```python
# how to find the table
table = driver.find_element(By.CLASS_NAME, 'grilla')
df = pd.read_html(table.get_attribute('outerHTML'))[0]
df
```

```python
# get the table, append to a list and follow to the next page

dfs = []

for i in range(0,424):
    table = driver.find_element(By.CLASS_NAME, 'grilla')
    df = pd.read_html(table.get_attribute('outerHTML'))[0]
    dfs.append(df)
    next_link = driver.find_element(By.XPATH, "//span[contains(@class, 'Label')]/parent::td/following-sibling::td/a")
    next_link.click()
    time.sleep(8)
```

```python
# join the dataframes
final_df = pd.concat(dfs)
final_df.info()
```

```python
# clean the data
final_df2 = final_df.drop(columns = ['Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11'])
final_df2

final_df2.to_csv('file_name.csv')
```
